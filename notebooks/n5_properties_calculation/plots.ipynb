```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Load the CSV file
file_path = '../../data/d5_properties_calculation/properties.csv'
properties_df = pd.read_csv(file_path)
```

```python
# Helper function to save plots
def save_plot(fig, filename):
    fig.savefig(f'../../plots/p5_properties_calculation/{filename}', bbox_inches='tight')
```

```python
# Plot 1: Count of Metals vs Non-Metals (Pie Chart)
fig1, ax1 = plt.subplots()
counts = properties_df['Is_Metal'].value_counts()
properties_df['Is_Metal'].value_counts().plot.pie(autopct='%1.1f%%', ax=ax1, labels=[f'Non-Metal ({counts[0]})', f'Metal ({counts[1]})'])
ax1.set_ylabel('')
ax1.set_title('Distribution of Metals vs Non-Metals')
save_plot(fig1, 'metals_vs_non_metals_pie_chart.png')
```
